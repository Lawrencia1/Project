```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #this can also be written as "from matplotlib import pyplot as plt"
import numpy as np
```

QUESTION 3 - Analyse the data to know which regions and what type of property should be invested in

a. Detached Houses

```python
#Step 1: Read in specific data for each house type and save into a named file 
detached = pd.read_excel("house_data_bricks_and_acres.xlsx", sheet_name= "1b", skiprows= (0,1))
detached
```

```python
#Step 2: Observe the data type
detached.head
#detached.shape
```

```python
#Step 3: Select needed columns from table
dh_table = detached[["Name", "Year ending Sep 2023"]]
dh_table
```

```python
#Step 4: Plot graph to illustrate price distribution across regions for the house type
dh_table.plot(kind = 'bar', title= "Prices of detached houses per area in Sep 2023", color="green", x= "Name", y="Year ending Sep 2023")
plt.xlabel("Location")
plt.ylabel("House Prices")
```
